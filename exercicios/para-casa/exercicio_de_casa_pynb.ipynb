```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('data_sample_SP.csv', sep=';')
df.head()
```

```python

df.groupby('ESTACAO')[['PRECIPITAO TOTAL. HORARIO (mm)']].mean().plot(color= 'purple', figsize=(10, 5))
plt.xlabel('Estação')
plt.ylabel('Precipitação total (mm)')
plt.title('Média da Precipitação por Estação')
```

```python
df.groupby('ESTACAO')[['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)']].mean().plot(color= 'orange', figsize=(10, 5))
plt.xlabel('TEMPERATURA DO AR - BULBO SECO. HORARIA (C)')
plt.ylabel('Frequência')
plt.title('Temperatura do AR')
```
